```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
fmri = sns.load_dataset("fmri")
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```
